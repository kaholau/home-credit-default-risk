```python
%load_ext autoreload
%autoreload 2

import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression

import gc
import myutil
```

```python
class lgrWrapper():
    def __init__(self):
        return
    
    def init_model(self):
        self.clf = LogisticRegression(C = 0.0001)     
        return    
    def __repr__(self):
        return 'logReg'

    def train(self, trn_x, trn_y):
        self.clf.fit(trn_x, trn_y)

    def predict(self, x):
        result = self.clf.predict_proba(x)[:, 1]
        return result
    
    def get_coef(self):
        return self.clf.coef_.reshape(-1,1)

lgr = lgrWrapper()
```

```python
data, test, y = myutil.get_train_test_label()
```

```python
oof_preds, test_preds, importances, folds = myutil.cross_validation_undersample(data, test, y, lgr)
```

```python
title = '{}_undersample'.format(lgr)
myutil.report(test_preds, folds, importances, data, y, oof_preds, title)
```
